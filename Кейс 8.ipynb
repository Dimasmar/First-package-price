```python
import pandas as pd
import scipy.stats as st
import numpy as np
import plotly.express as px
pd.options.mode.copy_on_write = False
```

```python
df = pd.read_csv("F:\Кейс 8/first_package_price_test.csv")
```

```python
dft = df.groupby("id_user").agg(
    t1=("split_group", "nunique"),
    t2=("gender", "nunique"),
    t3=("os", "nunique"),
    t4=("browser", "nunique"),
    t5=("age_group", "nunique"),
    t6=("device", "nunique"),
    t7=("traffic_type", "nunique"),
    t8=("country_group", "nunique")
)
```

```python
most_relevant_browsers = df["browser"].to_frame().groupby("browser", as_index=False).agg(users=("browser", "count")).sort_values(by="users", ascending=False)
most_relevant_browsers["percentage"] = most_relevant_browsers["users"]/df["id_user"].size

most_relevant_os = df["os"].to_frame().groupby("os", as_index=False).agg(users=("os", "count")).sort_values(by="users", ascending=False)
most_relevant_os["percentage"] = most_relevant_os["users"]/df["id_user"].size

df.loc[df["browser"].isin(most_relevant_browsers[most_relevant_browsers["percentage"] <= 0.0005]["browser"].to_list()), 'browser'] = 'Other'
df.loc[df["os"].isin(most_relevant_os[most_relevant_os["percentage"] <= 0.0005]["os"].to_list()), 'os'] = 'Other'
```

```python
df["has_paid"] = df["amount_1d"] > 0
for i in [1, 3, 7, 14, 30]:
    df["remains_" + str(i) + "d"] = df["spend_" + str(i) + "d"] / (130 * df["split_group"] * (df["amount_" + str(i) + "d"] / (df["split_group"] * 17 + 3)) + 20)
# df[df["amount_1d"] > df["amount_3d"]]
```

```python
grouped_df = df.groupby(["gender", "country_group", "age_group", "traffic_type", "browser"], as_index=False).agg(
    users=("id_user", "count"),
    paid_users=("has_paid", "sum"),
    ARPU=("amount_1d", "mean")
#     CR_to_paid=("amount_1d", "")
).sort_values("users", ascending=False)
grouped_df
# , "split_group"
```

```python
grouped_df["CR_to_paid"] = grouped_df["paid_users"]/grouped_df["users"]
```

```python
split_df = df.groupby(["gender", "country_group", "age_group", "traffic_type", "browser", "split_group"], as_index=False).agg(
    users=("id_user", "count"),
    paid_users=("has_paid", "sum"),
    ARPU=("amount_1d", "mean"),
    activity=("remains_1d", "sum")
#     CR_to_paid=("amount_1d", "")
).sort_values("users", ascending=False)
split_df
```

```python
final_df = split_df.merge(grouped_df[grouped_df["ARPU"] > 0], how='inner', on=["gender", "country_group", "age_group", "traffic_type", "browser"])
final_df
```

```python
final_df["activity"] = final_df["activity"]/final_df["paid_users_x"]
final_df
```
